### Definition

Eine Aggregatfunktion fasst mehrere Zeilen zu einem Ergebniswert zusammen.

Typische Aggregatfunktionen:

| Funktion | Bedeutung |
|---|---|
| `SUM(spalte)` | Summe der Werte |
| `COUNT(*)` | Anzahl der Ergebniszeilen |
| `COUNT(spalte)` | Anzahl der nicht-NULL-Werte in dieser Spalte |
| `AVG(spalte)` | Durchschnittswert |
| `MIN(spalte)` | kleinster Wert |
| `MAX(spalte)` | größter Wert |

## Gruppierung

Der Befehl `GROUP BY` wird verwendet, um mehrere Datensätze zu Gruppen zusammenzufassen.

Dadurch können Aggregatfunktionen für jede Gruppe getrennt berechnet werden.

```sql
SELECT kategorie, SUM(menge)
FROM bestellungen
GROUP BY kategorie; 
```

Beispiel:

```SQL
SELECT kategorie, MIN(einzelpreis),MAX(einzelpreis)
FROM bestellungen
GROUP BY kategorie
HAVING MIN(einzelpreis) >5
ORDER BY MIN(einzelpreis)
```

Ergebnis: Für jede Kategorie wird eine eigene Summe berechnet.

Wichtige Regel bei GROUP BY:

Wenn eine Aggregatfunktion verwendet wird, dürfen zusätzlich nur Spalten ausgegeben werden:

* die selbst aggregiert werden
    oder
* die in GROUP BY stehen.

Richtig:

```sql
SELECT kategorie, COUNT(*)
FROM bestellungen
GROUP BY kategorie;
````

Falsch:
```sql
SELECT artikel, kategorie, COUNT(*)
FROM bestellungen
GROUP BY kategorie;
````

Der Befehl HAVING wird verwendet, um Gruppen nach einer Aggregatfunktion zu filtern.

HAVING arbeitet also nicht auf einzelnen Zeilen, sondern auf bereits gebildeten Gruppen. 

## Reihenfolge der SQL-Befehle bei SELECT

Die Befehle werden in dieser Reihenfolge geschrieben:

```sql
SELECT ...
FROM ...
WHERE ...
GROUP BY ...
HAVING ...
ORDER BY ...
````
Warum genau diese Reihenfolge?

Die Reihenfolge entspricht der Verarbeitung der Datenbank.

SQL arbeitet die Daten Schritt für Schritt ab.

⸻

Reihenfolge der Datenverarbeitung

1. FROM

Zuerst werden die Tabellen ausgewählt.

2. WHERE

Danach werden einzelne Zeilen gefiltert.

3. GROUP BY

Nun werden die verbleibenden Zeilen gruppiert.

4. Aggregatfunktionen

Innerhalb jeder Gruppe werden nun Funktionen berechnet:

5. HAVING

Danach können Gruppen gefiltert werden.

6. SELECT

Jetzt werden die gewünschten Spalten ausgewählt.

7. ORDER BY

Ganz am Ende wird sortiert.

### Beispiel

```sql
SELECT
    kategorie,
    SUM(menge) AS gesamtmenge
FROM bestellungen
WHERE bezahlt = true
GROUP BY kategorie
HAVING SUM(menge) > 3
ORDER BY gesamtmenge DESC;
```

Ablauf dieses Beispiels

1. Tabelle bestellungen laden
2. Nur bezahlte Bestellungen auswählen (WHERE)
3. Nach Kategorie gruppieren (GROUP BY)
4. Summe je Kategorie berechnen (SUM)
5. Gruppen mit Summe > 3 behalten (HAVING)
6. Ergebnis ausgeben (SELECT)
7. Nach Gesamtmenge sortieren (ORDER BY)
