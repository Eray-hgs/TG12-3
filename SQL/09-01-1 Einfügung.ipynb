### Unterbefehele von SELECT

- FROM -> Auswahl von Tabellen
- WHERE -> Eingrezen der ausgzugebeben Zeilen
- GROUP BY 
- ORDER BY -> Sortieren der Zeilen
    - DESC -> absteigende Sortierenung
    - ASC -> aufsteigende Sortierung


### Logische Verknüpfungen
- LIKE -> Vergleich Text mit Muster, z.B name LIKE "ma%"
- AND -> Abfrage mit mehreren Bedingungen, z.B name LIKE "ma%" AND status LIKE "friedlich"
- OR -> Abfrage mit mehreren Bedingungen, aber nur eine muss stimmen, z.B name LIKE "ma%" OR status LIKE "friedlich"
- COUNT(*)
- SUM()
- AVG()
- IS NULL -> Abfrage, ob ein Attribut keinen Wert hat
- IN -> Abfrage, ob der Wert des Attributes in einer Liste vorkommt
>
- UPDATE -> ändert Daten von Zeilen
- DELETE -> löscht Zeilen
