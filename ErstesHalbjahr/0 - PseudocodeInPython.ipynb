```python
def findeLagerplatz(self, pArtikelname):
    idxGefunderLp : int
    idx : int
    name : str
    frei : bool
    gefunden : bool

    idxGefunderLp = -1
    gefunden = False
    idx=0

    for idx in range (0,self.aAnz,1):
        name =self.derLagerplatz[idx].gibArtikelname()
        if name == pArtikelname:
            gefunden == True
            idxGefunderLp = idx
            self.dieGui.melden(f"Lagerplatz {idxGefunderLp} gefunden")
    
    if gefunden == False:
        for idx in range(0,self.aAnz,1):
            frei = self.derLagerplatz[idx].istFrei()
            if frei == True:
                gefunden = True
                idxGefunderLp = idx
                self.dieGui.melden("Wahre wird eingelagert")
            
    if gefunden == False:
        self.dieGui.melden("Lager voll, einlagern nicht m√∂glich")

    return idxGefunderLp
```
