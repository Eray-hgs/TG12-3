# Freierfall ohne Luftwiederstand



```python
import math
import matplotlib.pyplot as plt
import numpy as np

t = 0
g = 9.81
m = 80
v = 0
s = 0
dt = 0.01

tr =[]
vr =[]
sr =[]
sg =[]

for i in range(0, 100):
    t   = t + dt
    a   = g
    v   = v + a*dt
    s   = s + v*dt
    s2  = 0.5*g*t**2
    
    tr.append(t)
    vr.append(v)
    sr.append(s)
    sg.append(s2)


plt.plot(tr, sr, color='red')
plt.plot(tr, sg, color='blue')
plt.xlabel('t in s')
plt.ylabel('s in m')
plt.title('Freier Fall')
plt.grid()
plt.show() 



```
