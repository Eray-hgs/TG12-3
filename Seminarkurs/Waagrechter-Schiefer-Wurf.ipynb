# Schiefer Wurf

### Teilaufgabe 1

```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Schiefer Wurf
g = 9.81  # Erdbeschleunigung in m/s^2
v = 30  # Anfangsgeschwindigkeit in m/s
t = 0  # Zeit in Sekunden
dt= 0.01  # Zeitintervall in Sekunden

winkel = 30  # Abwurfwinkel in Grad

vx = v * math.cos(math.radians(winkel))
vy = v * math.sin(math.radians(winkel))

xR = []
yR = []

for i in range (0, 310):
    t = t + dt
    x = vx * t
    y = vy * t - 1/2 * g * t**2
    xR.append(x)
    yR.append(y)

plt.plot(xR, yR)
plt.title("Schiefer Wurf")
plt.xlabel("x in m")
plt.ylabel("y in m")
plt.ylim(bottom=0)
plt.grid()
plt.show() 
```

### Teilaufgabe 2 

```python
import math
import numpy as np
import matplotlib.pyplot as plt

# Schiefer Wurf
g  = 9.81  # Erdbeschleunigung in m/s^2
v0 = 15  # Anfangsgeschwindigkeit in m/s
t  = 0  # Zeit in Sekunden
dt = 0.001  # Zeitintervall in Sekunden

yi = 0
xi = 0
y  = 0
x  = 0

winkel = 60  # Abwurfwinkel in Grad

vx = v0 * math.cos(math.radians(winkel))
vy = v0 * math.sin(math.radians(winkel))
vi = vy

xR1 = [0]
yR1 = [0]
xR2 = [0]
yR2 = [0]

for i in range (0, 6000):
    t = t + dt
    x = vx * t
    y = vy * t - 1/2 * g * t**2

    vi = vi + (-g) * dt
    yi = yi + vi * dt 
    xi = xi + vx * t
    
    try :
        if y < 0:
            break
    except:
        pass

    xR1.append(x)
    yR1.append(y)
    xR2.append(xi)
    yR2.append(yi)
  

plt.plot(xR, yR)
plt.plot(xR1, yR1,color='red')
plt.title("Schiefer Wurf")
plt.xlabel("x in m")
plt.ylabel("y in m")
#plt.ylim(bottom=0)
plt.grid()
plt.show()

```
