```python
import math
import numpy as np
from matplotlib import pyplot as plt

g = 9.81
y = 100
t = 0

while y >= 0:
    print("t:", t, " y:", y)
    t += 0.1
    y = 100 - 0.5 * g * (t**2)
    
t = np.arange(0, t, 0.1)
y = 100 - 0.5 * g * (t**2)
plt.plot(t, y, color='red')
plt.xlabel('x')
plt.ylabel('y')
plt.title('Plot von y = x^2')
plt.grid()
plt.show()  

```
